## 1️⃣ Color Codes: The Language of Digital Color

Colors on screens are defined using **color codes** that computers can understand.

### 💡 Common Formats

- **Hex Codes**: `#RRGGBB`  
  Uses base-16 (0–9, A–F)  
  Examples: `#000000` (Black), `#FFFFFF` (White), `#FF5733` (Reddish-orange)

- **RGB**: `rgb(R, G, B)`  
  Each value: 0–255  
  Example: `rgb(255, 87, 51)` = `#FF5733`

- **RGBA**: `rgba(R, G, B, A)`  
  Adds transparency with `A` from 0 (transparent) to 1 (opaque)  
  Example: `rgba(255, 87, 51, 0.5)`

| Color | Hex | RGB | RGBA |
|-------|-----|-----|------|
| Red | `#FF0000` | `rgb(255, 0, 0)` | `rgba(255, 0, 0, 1)` |
| Blue | `#0000FF` | `rgb(0, 0, 255)` | `rgba(0, 0, 255, 0.3)` |
| White | `#FFFFFF` | `rgb(255, 255, 255)` | `rgba(255, 255, 255, 1)` |

## 🗂️ 3️⃣ Image File Formats

| Format | Best Use | Transparency | Compression |
|--------|----------|--------------|-------------|
| **JPG** | Photos | ❌ No | Lossy |
| **PNG** | Logos, graphics | ✅ Yes | Lossless |
| **GIF** | Animations | ✅ Yes (limited) | Lossless (limited) |
| **SVG** | Icons, vectors | ✅ Yes | Vector |

**Summary:**

- **JPG**: Great for photos, small size, loses some detail.  
- **PNG**: High-quality, supports transparency.  
- **GIF**: For simple animations, limited colors.  
- **SVG**: Scales perfectly, used for clean web graphics.

## 🔢 4️⃣ How Images Store Color

Each pixel stores:

- R (Red): 0–255  
- G (Green): 0–255  
- B (Blue): 0–255  
- A (Alpha): 0.0–1.0 (optional)

### 🧠 Pixel Example

- RGB: `76, 175, 80`  
- Hex: `#4CAF50`  
- RGBA: `rgba(76, 175, 80, 1)`

## 🔐 5️⃣ Base64: Turning Binary into Text

### 🔍 What Is Base64?

A way to **encode binary data** into text using 64 characters:  
`A–Z`, `a–z`, `0–9`, `+`, `/`, with `=` used as padding.

### 📦 Why Use It?

Base64 lets us send images, files, or credentials in **text-only systems** (like emails, URLs, JSON).

### 🧠 How It Works

1. Convert binary to 6-bit chunks  
2. Map each chunk to a Base64 character

| Binary | Base64 |
|--------|--------|
| 000000 | A |
| 000001 | B |
| ...    | ... |
| 111111 | / |

### 🔄 Example: "Hi"

- Binary: `01001000 01101001`  
- 6-bit Chunks: `010010 000110 100100`  
- Decimal: 18, 6, 36  
- Base64: **SGk=**

## 🛠️ Real-World Base64 Uses

- Embed images/fonts in HTML or CSS  
- Attach files in emails  
- Pass data in APIs  
- Encode credentials (Basic Auth)
